```python
#   #   #    #   #   #
# 0. Insert Path TXT #
#   #   #    #   #   #

# Insert TXT file in HTML format for further manipulation
# Change Number of Path Ending in x with 1-24 - "...\\knesset_X.txt" for Corresponding Year of Knesset

file_temp = "C:\\...\\knesset_1.txt"
```

# Knesset Members - Year Of Birth

```python
#   #   #   #   #   #
# A. Year Of Birth  #
#   #   #   #   #   #


#
# 2.
#

# Open TXT file in HTML format
# Decode with utf-8 encoding

with open(file_temp, "r", encoding='utf-8') as f:
    
    # Convert decoded TXT file to STRING type
    decoded_txt = f.read()

#    
# 3.
#

# import Request module

import re

# Extract INITIAL: Dates of Knesset Members 
# ANYTHING between: text:dateLiveOrDead"> AND </div>
d = decoded_txt
long = re.findall(r'text:dateLiveOrDead">(.*?)</div>', d)

#
# 4.
#

# Take LONG list and KEEP only REAL DATES (where String length is more than 3, e.g. THOUSANDS and up)

dates_correct_length = [item for item in long if (len(item)>3)]

#
# 5.
#

# Extract Birth dates - Dirty, Only Ending 

for i in range(len(dates_correct_length)):
        dates_correct_length[i] = dates_correct_length[i][-5:]

#
# 6.
#

# Clean Dates Strings

import re 

# JOIN List as STRING 
dcl = ''.join(dates_correct_length)

# Extract Only Year of Birth INT as STRING. (Dropping all Special Characters)
dob_str = re.findall(r"\d\d\d\d", dcl)

# Convert Itens in List from STRING to INT
dob = list(map(int, dob_str))

##
## CHECK - Visual Test 
##

countMembers = len(dob)
strCount = str(countMembers)

print("Success!\n")
print(dob)
print("\nThere are - "+ strCount + " Members this Year.")

##
##  pkClean -> final list "partial knesset or not"
##  dob -> final list "Year of Birth"
##  names_mid4 -> final list "Full Names of Knesset Members"
##
```
